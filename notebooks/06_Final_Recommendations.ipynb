```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../top_40_priority_customers.csv")
```

```python
df.head()
```

```python
df["segment"].value_counts()
```

```python
df["segment"].value_counts().plot(
    kind="bar",
    color="red",
    figsize=(6,4)
)

plt.title("Customer Segment Distribution")
plt.ylabel("Number of Customers")
plt.show()
```

```python
df["industry"].value_counts()
```

```python
df["industry"].value_counts().plot(
    kind="bar",
    figsize=(8,5),
    color="orange"
)

plt.title("Industry Distribution")
plt.show()
```

```python
df["region"].value_counts()
```

```python
df["region"].value_counts().plot(
    kind="bar",
    color="green",
    figsize=(7,4)
)

plt.title("Region Distribution")
plt.show()
```
