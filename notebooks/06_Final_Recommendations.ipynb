```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../northbeam_customer_health_scores.csv")
```

```python
top_40 = df.nsmallest(40, "customer_health_score")

top_40
```

```python
top_40["segment"].value_counts()
```

```python
top_40["segment"].value_counts().plot(
    kind="bar",
    color="red",
    figsize=(6,4)
)

plt.title("Customer Segment Distribution")
plt.ylabel("Number of Customers")
plt.show()
```

```python
top_40["industry"].value_counts()
```

```python
top_40["industry"].value_counts().plot(
    kind="bar",
    figsize=(8,5),
    color="orange"
)

plt.title("Industry Distribution")
plt.show()
```

```python
top_40["region"].value_counts()
```

```python
top_40["region"].value_counts().plot(
    kind="bar",
    color="green",
    figsize=(7,4)
)

plt.title("Region Distribution")
plt.show()
```
