```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../northbeam_accounts_clean.csv")
```

```python
df_fe = df.copy()
```

```python
df_fe["signup_date"] = pd.to_datetime(df_fe["signup_date"])

today = pd.Timestamp.today()

df_fe["customer_tenure_days"] = (
    today - df_fe["signup_date"]
).dt.days
```

```python
df_fe["seat_utilization"] = (
    df_fe["seats_active"] /
    df_fe["seats_licensed"]
)
```

```python
df_fe["seat_utilization"] *= 100
```

```python
mau_cols = [
    "mau_m5",
    "mau_m4",
    "mau_m3",
    "mau_m2",
    "mau_m1",
    "mau_m0"
]

df_fe["avg_mau"] = df_fe[mau_cols].mean(axis=1)
```

```python
df_fe["usage_trend"] = (
    df_fe["mau_m0"] -
    df_fe["mau_m5"]
)
```

```python
df_fe["usage_trend"]
```

```python
df_fe["contract_end_date"] = pd.to_datetime(
    df_fe["contract_end_date"]
)

df_fe["days_to_renewal"] = (
    df_fe["contract_end_date"] -
    today
).dt.days
```

```python
df_fe["high_support"] = np.where(
    df_fe["support_tickets_90d"] >= 10,
    1,
    0
)
```

```python
df_fe["has_escalation"] = np.where(
    df_fe["escalations_90d"] > 0,
    1,
    0
)
```

```python
df_fe.to_csv(
    "../northbeam_feature_engineered.csv",
    index=False
)
```
